## nettoyer, standardiser et préparer les données pour l’analyse & Power BI

Charger le CSV propre

```python
import pandas as pd

path = "../data/processed/cta_ridership_by_route_dailyy.csv"
df = pd.read_csv(path)

df.head()
```

## Nettoyage & typage (propre)

```python
# Conversion des types
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["route"] = df["route"].str.strip()
df["rides"] = pd.to_numeric(df["rides"], errors="coerce")
print(df.dtypes)

```

```python
# Suppression des lignes invalides
df = df.dropna(subset=["date", "route", "rides"])
print(df.info())
```

```python
# Sauvegarde finale (dataset clean)
output_path = "../data/processed/cta_ridership_by_route_daily_clean.csv"
df.to_csv(output_path, index=False)

output_path
```
